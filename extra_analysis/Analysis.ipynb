# Short Data Visual Analysis
First we import necessary libraries

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 1000)
```

###  Next we load our csv file from Athena

```python
df=pd.read_csv('results.csv')
df
```

### Next stop to create a new column to calculate percent change or stock hourly volatility

```python
df['pct']=df['high_price'].pct_change()
```

### Further feature processing to add and group by our ticker names in order to find the frequency  of hourly identical price occurences

```python
df['same']=df['ticker'].shift()== df['ticker']
df1 = df[df.same != False]

df2 = df1
df2['dupl'] = df2['high_price'].duplicated()
```

### First Graph shows stocks with most occurences of  price reaching same value  (or staying same) within one hour

```python
df2.groupby(['ticker'])['dupl'].sum().sort_values().plot(kind='barh',figsize=(7,8),color = 'lightgreen')
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'
```

### Next, we graph percent change for each stock. We see that stock with largest negative percent change  is BYND and SQ, while stocks with largest positive pct change are DDOG and TTD.

Another interesting observation: Most stock volatility happens ealier in the day ( around 10-11 am) and by the end of the day ( mostly positive)

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = df1, x='hour', y='pct', hue='ticker', legend = 'brief')
```

### Basic visual price representation through the day 

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = df, x='hour', y='high_price', hue='ticker', legend = 'brief')
```

### Barplot, representing number of hour occurences. No hour has exactly 10 value counts, meaning that every hour had at least one occurence of same company reaching same stock price within the hourly time limits

```python
df['hour'].value_counts().plot(kind='bar',figsize=(7,8),style = 'fivethirtyeight')
```

### Histogram dist of the percent change among all of the 10 companies on May 14th 2020

```python
df1['pct'].plot(kind='hist',color='lightblue',edgecolor='gray',figsize=(7,7))
plt.title('Histogram', size=16)
plt.xlabel('PercentChange', size=18)
plt.ylabel('Frequency', size=18)
```
